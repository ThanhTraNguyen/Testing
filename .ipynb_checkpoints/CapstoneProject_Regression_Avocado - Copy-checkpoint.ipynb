# Question 1: Predict AveragePrice


```python
df.columns
```

```python
# Create a data set with only interested variables
df_regress = df.drop(['Date', 'AveragePrice', '4046', '4770', '4225', 'Small Bags', 'Large Bags', 'XLarge Bags', 'season'], axis=1)
df_regress.head()
```

```python
# Log-transforming variables with skew distributions.
for i in ['Total Volume', 'Total Bags']:
  df_regress[i] = np.log(df_regress[i]+1)

df_regress.head()
```

```python
#pd.get_dummies(df_regress)
```

```python
# Convert string data to number
# Type: Assuming that there are only 2 type of avocado (no new value of type variable in future) -> set of avocado type in training set will be universal -> can convert the whole data set before spliting
# Region: Assuming that there are many cities that maybe some cities in future are not included tranining set -> spliting first and then converting later to avoid leaking information.

Label_type = LabelEncoder()
df_regress['type']=Label_type.fit_transform(df_regress['type'])
df_regress.head()
```

```python
# variable month here is also a non-numertic variable. Every month is a time point in a year

OneHotEncoder_month = OneHotEncoder(handle_unknown='ignore', sparse=False)
OneHotEncoder_month = OneHotEncoder_month.fit(df_regress[['month']])
month = pd.DataFrame(OneHotEncoder_month.transform(df_regress[['month']]))
month.columns = OneHotEncoder_month.get_feature_names_out(['month'])
month.index = df_regress.index
df_regress = pd.concat([df_regress.drop(columns = ['month']), month], axis = 1)
df_regress.head()
```

```python
# Train-test split
X=df_regress
y=df['AveragePrice']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Build OneHotEncoder model from converting region variable in training set

OneHotEncoder_region = OneHotEncoder(handle_unknown='ignore', sparse=False)
OneHotEncoder_region = OneHotEncoder_region.fit(X_train[['region']])
region = pd.DataFrame(OneHotEncoder_region.transform(X_train[['region']]))
region.columns = OneHotEncoder_region.get_feature_names_out(['region'])
region.index = X_train.index
X_train = pd.concat([X_train.drop(columns = ['region']), region], axis = 1)
```

```python
# All feature
all_features = X_train.columns
print(all_features)
X_train.head()
```

```python
# Use OneHotEncode_Region to transform X_test data

region = pd.DataFrame(OneHotEncoder_region.transform(X_test[['region']]))
region.columns = OneHotEncoder_region.get_feature_names_out(['region'])
region.index = X_test.index
X_test = pd.concat([X_test.drop(columns = ['region']), region], axis = 1)
```

```python
X_test.head(2)
```

```python
# Scale data
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Check by using lazy predict
# clf = LazyRegressor(verbose=0, ignore_warnings = True, custom_metric = None)
# models, predictions=clf.fit(X_train,X_test,y_train,y_test)
# models
# Model có R_squared cao hoặc/và thời gian fit thấp: ExtraTreesRegressor, XGBRegressor
```

```python
# Function to test models, using cross validatio (double-check with Lazy predict)
def train_model(models, X_train, y_train, X_test, y_test):
    """5-fold cross validations.
    Args: models, X_train, y_train, X_test, y_test
    Returns: A dataframe with the algoithm's name, Average R2 score (train & test) across 5-fold cross validations.
    """
    training_model = pd.DataFrame(columns = ['Model', 'Avg_R2Score_Train', 'Avg_R2Score_Test'])

    for model in models:
        model.fit(X_train, y_train)
        avg_score_train = cross_val_score(model, X_train, y_train, scoring ='r2', cv = 5).mean()
        avg_score_test = cross_val_score(model, X_test, y_test, scoring ='r2', cv = 5).mean()
        score = pd.DataFrame([[str(model).split('(')[0], avg_score_train, avg_score_test]], columns = ['Model', 'Avg_R2Score_Train', 'Avg_R2Score_Test'])

        training_model = pd.concat([training_model, score], ignore_index=True, axis = 0)
    return training_model.sort_values(by=['Avg_R2Score_Test'])
```

```python
# Models
models = [LinearRegression(), KNeighborsRegressor(), RandomForestRegressor(), XGBRegressor(), SVR(), GradientBoostingRegressor(), ExtraTreesRegressor(), RandomForestRegressor()]
model_scores = train_model(models, X_train, y_train, X_test, y_test)
model_scores
```

```python
# Model ExtraTreesRegressor is suggested in both Lazy-predict & cross_validation, accuracy 92%
```

```python
# Build ExtraTreesRegressor_model
reg= ExtraTreesRegressor()
reg.fit(X_train,y_train)
```

```python
# Use ExtraTreesRegressor_model to check the importance of variables in the dataset
# -> Type, year, total bags

feat_importances = pd.Series(reg.feature_importances_, index=all_features)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
# Use ExtraTreesRegressor_model to predict test set
df_test_predict = pd.DataFrame({'actual': y_test, 'predict': reg.predict(X_test)})
df_test_predict.head()
```

```python
# Predict test
future_forecast = stepwise_model.predict(n_periods = len(test)) # số khoảng thời gian cần dự đoán
future_forecast = pd.DataFrame(future_forecast, index = test.index, columns=['Prediction'])
future_forecast.head()
```

```python
# Error approximates 20% of the average price, mae is quite high, need to be improve
mae/mean_price
```

## Prophet

```python
df_LA_organic.head()
```

```python
# Train-test
df_LA_organic = df_LA_organic.reset_index(level=0)
df_LA_organic.columns = ["ds", "y"]

train = df_LA_organic.iloc[:round(len(df_LA_organic)*0.7)]
test = df_LA_organic.iloc[round(len(df_LA_organic)*0.7):]
```

```python
model = Prophet(yearly_seasonality = True,
                daily_seasonality=False, weekly_seasonality=False)
model.fit(train)
```

```python
train.tail()
```

```python
test.tail()
```

```python
# Predict new values test and next 3 years
months = pd.date_range('2017-04-01','2021-04-01', freq = 'MS').strftime("%Y-%m-%d").tolist()
future = pd.DataFrame(months)
future.columns = ['ds']
future['ds'] = pd.to_datetime(future['ds'])
forecast = model.predict(future)
forecast[['ds', 'yhat']].head()
```

```python
# Calculate MAE between expected and predicted values for test
y_test = test['y'].values
y_pred = forecast['yhat'].values[:len(test)]
mae = mean_absolute_error(y_test, y_pred)
print('MAE: %.3f' % mae)
```

```python
average_price = df_LA_organic.y.mean()
average_price
```

```python
mae/average_price
```

```python
# Error approximates 13% of the price. Comparing to Arima, Prophet is a suitable model for this data set
```
